```python
# подключаем библиотеки
import sklearn
import numpy as np
import pandas as pd
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
plt.scatter(data.BMI, data.Age)
plt.xlabel('BMI')
plt.ylabel('Age')
```

```python
plt.figure(figsize=(8, 6))

# Визуализация данных с диабетом (Outcome=1) в красном цвете
plt.scatter(data[data['Outcome'] == 1]['BMI'], data[data['Outcome'] == 1]['Age'], color='red', label='Diabetes')

# Визуализация данных без диабета (Outcome=0) в синем цвете
plt.scatter(data[data['Outcome'] == 0]['BMI'], data[data['Outcome'] == 0]['Age'], color='blue', label='No Diabetes')

plt.xlabel('BMI', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.legend()
plt.show()
```

```python
#смотримм, нужна ли нормализация
data[['BMI', 'Age']].head()
#не нужна, данные лежат, в принципе, в одном масштабе
```

```python
km = KMeans(n_clusters=2)
y_pred = km.fit_predict(data[['BMI', 'Age']])
y_pred
```

```python
km.cluster_centers_
```

```python
#визуализация кластеров
plt.figure(figsize=(9, 6))

plt.scatter(data[y_pred==0]['BMI'], data[y_pred==0]['Age'], color='red', label='Cluster 1')
plt.scatter(data[y_pred==1]['BMI'], data[y_pred==1]['Age'], color='blue', label='Cluster 2')

plt.xlabel("BMI", fontsize=14)
plt.ylabel("Age", fontsize=14)
plt.legend(loc="upper left", fontsize=12)
plt.show()
```

```python
#Иерархическая дендрограмма
```

```python
data=pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
y=data['Outcome']
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

fig = plt.figure(figsize=(16, 50))
fig.patch.set_facecolor('white')

# Реализация иерархической кластеризации при помощи функции linkage
mergings = linkage(data[['BMI', 'Age']], method='ward')

# Построение дендрограммы. Разными цветами выделены автоматически определенные кластеры
R = dendrogram(mergings, labels=[y[i] for i in y], orientation = 'left', leaf_font_size = 4)

# Отображение дендрограммы
plt.show()
```
